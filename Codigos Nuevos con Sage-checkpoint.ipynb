```python
#Ahora haremos el mismo código de antes solo que comprobando si el código dual correspondiente a cada código es también AMDS, haciendo al código original NMDS


#Derivada de Hasse-Schmidt

def HasseSchmidtDer(n,y,t):
    return binomial(n,t)*y^(n-t)

#El vector a extender en cualquier punto, salvo el infinito, usando el programa anterior

def vectorextension(k,y,t):
    return [[HasseSchmidtDer(n,y,t)] for n in range(k)]

#El vector a extender en el punto del infinito, este tiene que ser hecho por separado, ya que no es sencillo programar un cambio de coordenadas en SAGE.

def vectorextensioninf(k,t):
    return [[kronecker_delta(k-n-1,t)] for n in range(k)]

def posibles_extensiones(q,k):
    s=[]
    for i in range(1,q+1):
        for j in range(1,k):
            s.append([i,j])
    t=[Combinations(s,1).list()]
    for i in range(2,q+k):
        t.append(Combinations(s*i,i).list())
    return flatten(t,max_level=1)

#Probamos el generador de combinaciones

print(posibles_extensiones(2,3))
print()

#Sabiendo que funciona ahora hacemos un programa que determine los códigos AMDS y qué extensiones funcionan (trivialmente las que solo extienden por una columna son AMDS)

def extensiones_totales_NMDS(q,k):


    #Definimos la longitud de la lista de extensiones, una nueva tabla S el cuerpo Fq y su generador al que llamaremos x
    T=[]
    S=[]
    F=GF(q)
    x=F.gen()

    #Acto seguido definimos la matriz generadora de los códigos generalized Reed-Solomon añadiendo la columna de infinito

    G = codes.GeneralizedReedSolomonCode(F.list(),k)
    G = G.generator_matrix()

    row = vectorextensioninf(k,0)
    row = matrix(F,row)
    G = G.augment(row)

    #Definimos G_estable=G para poder reescribarla luego

    G_estable=G


    #Después extendemos matrices utilizando las indicaciones de la lista t (si los elementos de t son enteros entre 1 y q-1 entonces se aplica       la extension por la derivada que aparece en el segundo término, en caso contrario se deriva por el infinito)
    #Esto nos quiere decir que t tiene que ser de la forma t=[[1,2],[3,5],[-1,4],...]

    s=[]
    for i in range(1,q+1):
        for j in range(1,k):
            s.append([i,j])
    for j in range(1,q+k):
        print(j)
        T=Combinations(s*j,j).list()
        for i in range(len(T)):
            t=T[i]
            for el in t:
                if el[0] in ZZ and 1<=el[0] and el[0]<=q-1:
                    row = vectorextension(k,x^(el[0]),el[1])
                else:
                    row = vectorextensioninf(k,el[1])
                row = matrix(F,row)
                G = G.augment(row)

        #Terminamos calculando los datos del código generado por la nueva matriz G

            C=codes.LinearCode(G)
            C_dual=C.dual_code()

            d=C.minimum_distance()
            d_dual=C_dual.minimum_distance()
            n=C.length()
            k=C.dimension()

        #Si el código es AMDS lo incluímos en la lista final, y después cambiamos la matrix G por G_estable que definimos antes

            if d==n-k and d_dual==k:
                S.append([n,k,d,q,t])
                print(C.weight_enumerator())
                if n>=q+4:
                    print([n,k,d,q,t])
                    print(C.weight_enumerator())
                    print()
            G=G_estable

    return S

#Probamos el código

extensiones_totales_NMDS(5,4)
```
